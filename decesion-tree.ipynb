```python
from sklearn.tree import export_graphviz
import graphviz
from sklearn import tree
```

```python


import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from sklearn.tree import DecisionTreeClassifier 
from IPython.display import Image  
```

Reading Dataset

```python
df = pd.read_csv("../input/dtreepy/Iris.csv")
df.head()
fe = ['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']
```

forming data frame required for classifier

```python
x = df.iloc[:,[1,2,3,4]]
x.head()
```

```python
y = df.iloc[:,[5]]
y.head()
```

```python
Defining decision tree algorithm
```

```python
dtree=DecisionTreeClassifier()
dtree.fit(x,y)
```

Visualizing decision tree 

```python
data = tree.export_graphviz(dtree,feature_names=fe,out_file=None)
graph = graphviz.Source(data)
graph
```

Predicting type of species based on trained classifier

```python
pred = dtree.predict([[5,3,2,0.3]])
pred
```
